```python
#Using tensorflow 2.10
import pandas as pd
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
```

```python
adress = r'C:\Users\Lara\Downloads\house_prices.csv'
content = pd.read_csv(adress)
```

```python
content.head()
```

```python
square_footage = content.iloc[:, 5:6] #todas as linhas, coluna 5, não pega o lower
house_price = content.iloc[:, 2:3] #não pega o lower
```

```python
square_footage_training, square_footage_test, house_price_training, house_price_test = train_test_split(square_footage, house_price, test_size=0.3)
```

```python
def training_input_fn():
    dataset = tf.data.Dataset.from_tensor_slices(({'column': np.array(square_footage_training)}, np.array(house_price_training)))
    dataset = dataset.shuffle(buffer_size=len(square_footage_training))
    dataset = dataset.batch(32)
    return dataset
```

```python
def test_input_fn():
    dataset = tf.data.Dataset.from_tensor_slices(({'column': np.array(square_footage_test)}, np.array(house_price_test)))
    dataset = dataset.shuffle(buffer_size=len(house_price_test))
    dataset = dataset.batch(32)
    return dataset
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(1,), name='column'),  # Input layer
    tf.keras.layers.Dense(1)  # Linear output layer
])
```

```python
model.compile(optimizer='adam', loss='mse')
```

```python
model.fit(training_input_fn(), epochs=100, steps_per_epoch=len(square_footage_training) // 32)
```

```python
result_test = model.evaluate(x={'column': np.array(square_footage_test)}, y=np.array(house_price_test))
result_training = model.evaluate(x={'column': np.array(square_footage_training)}, y=np.array(house_price_training))
```
